## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# --- Cleaning titles dataset ---

# Drop rows with missing title or ID
titles_df.dropna(subset=['title', 'id'], inplace=True)

# Fill missing 'age_certification' with 'Unknown'
titles_df['age_certification'].fillna('Unknown', inplace=True)

# Fill missing numeric fields with median
numeric_cols = ['runtime', 'imdb_score', 'imdb_votes', 'tmdb_popularity', 'tmdb_score']
for col in numeric_cols:
    if col in titles_df.columns:
        titles_df[col].fillna(titles_df[col].median(), inplace=True)

# Fill missing 'seasons' with 0 (only relevant for shows)
titles_df['seasons'].fillna(0, inplace=True)

# Convert genres and production_countries to lists if they are stored as strings
import ast
import pandas as pd

# Safely evaluate only if the value is a string
for col in ['genres', 'production_countries']:
    titles_df[col] = titles_df[col].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)


# --- Cleaning credits dataset ---
# Drop rows with missing person name or ID
credits_df.dropna(subset=['name', 'id'], inplace=True)

# Fill missing character names with 'Unknown'
credits_df['character'].fillna('Unknown', inplace=True)

# --- Merge datasets for analysis ---
merged_df = pd.merge(titles_df, credits_df, on='id', how='left')

# Save the merged dataset to a new CSV file
merged_df.to_csv("merged_amazon_prime_dataset.csv", index=False)

print("Merged dataset saved as 'merged_amazon_prime_dataset.csv'")

# Preview cleaned and merged dataset
merged_df.head()
````

#### Chart - 14 - Correlation Heatmap

````python
# Correlation Heatmap visualization codee

# Select specific numerical columns for correlation
numeric_df = df[['runtime', 'imdb_score']].dropna()

# Compute the correlation matrix
correlation_matrix = numeric_df.corr()

# Plot the heatmap
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap - Runtime vs IMDb Score')
plt.show()
````

##### 1. Why did you pick the specific chart?

A correlation heatmap is perfect for quickly visualizing the strength and direction of relationships between numerical variables. It simplifies understanding which factors influence each other. In this case, it was chosen to examine how strongly runtime correlates with imdb_score.

##### 2. What is/are the insight(s) found from the chart?

The chart shows a very weak positive correlation (0.09) between runtime and IMDb score. This means there's almost no linear relationship between how long a movie or show is and how well it is rated, confirming that runtime does not significantly impact audience rating.

#### Chart - 15 - Pair Plot

````python
# Pair Plot visualization code
# Selecting numeric columns for pairplot
pairplot_data = df[['runtime', 'imdb_score', 'release_year']].dropna()

# Create the pairplot with axis titles
pairplot = sns.pairplot(pairplot_data, diag_kind='kde')

# Add a super title for the entire grid
pairplot.fig.suptitle("Pairwise Distribution of Runtime, IMDb Score, and Release Year", y=1.03)

# Set axis titles for each subplot
for i, row_var in enumerate(pairplot_data.columns):
    for j, col_var in enumerate(pairplot_data.columns):
        if i != len(pairplot_data.columns)-1:
            if j == 0:
                pairplot.axes[i][j].set_ylabel(row_var, fontsize=10)
        if i == len(pairplot_data.columns)-1:
            pairplot.axes[i][j].set_xlabel(col_var, fontsize=10)

plt.show()
````

##### 1. Why did you pick the specific chart?

A pairplot is ideal for exploring relationships between multiple numerical variables at once. It shows individual distributions on the diagonal and scatter plots for pairwise comparisons off-diagonal. This chart allows us to detect correlations, trends, and possible outliers among runtime, IMDb score, and release year in a single, compact view.

##### 2. What is/are the insight(s) found from the chart?

From the chart, we see that IMDb score and runtime show weak or no correlation, indicating that longer content doesn't guarantee better ratings. A strong trend is visible between release year and runtime, with newer content tending toward more consistent runtime lengths. IMDb scores appear normally distributed.
